```python
import re
```

```python
Clinton = df.iloc[6]['transcripts']
Trump = df.iloc[9]['transcripts']
```

```python
# Clear speeches from tags, wrong punctuation and double spaces
Clinton=re.sub("\[.*?\]","",Clinton)
Clinton=re.sub("The President."," [President] ",Clinton)
Clinton=re.sub("-Q."," [Q] ",Clinton)
Clinton=re.sub(" Q."," [Q] ",Clinton)
Clinton=re.sub("be: \( 1"," be ",Clinton)
Clinton=re.sub(" '","' ",Clinton)
Clinton=re.sub("  "," ",Clinton)
```

```python
Trump=re.sub("\(.*?\)"," ",Trump)
Trump=re.sub("be: \( 1"," be ",Trump)
Trump=re.sub("\[.*?\]","",Trump)
Trump=re.sub("– -",", ",Trump)
Trump=re.sub(" –",", ",Trump)
Trump=re.sub(" Q—"," [Q] ",Trump)
Trump=re.sub("—",", ",Trump)
Trump=re.sub("‘"," ' ",Trump)
Trump=re.sub(" Q "," [Q] ",Trump)
Trump=re.sub("  "," ",Trump)
```
